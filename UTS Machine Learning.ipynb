1) Machine Learning (ML) adalah cabang dari kecerdasan buatan (Artificial Intelligence/AI) yang fokus pada pengembangan algoritma dan model yang memungkinkan komputer untuk belajar dari data dan membuat keputusan atau prediksi tanpa diprogram secara eksplisit. Konsep utama dari machine learning adalah kemampuan sistem untuk memperbaiki kinerjanya berdasarkan pengalaman atau data baru.

Cara Kerja Machine Learning
Secara umum, proses machine learning melibatkan beberapa langkah kunci:
1. Pengumpulan Data : Data dikumpulkan dari berbagai sumber dan digunakan sebagai input untuk model machine learning.
2. Pra-pemrosesan Data : Data sering kali perlu dibersihkan dan diubah agar cocok untuk digunakan oleh model. Ini termasuk penanganan nilai yang hilang, normalisasi, dan transformasi data.
3. Pembagian Data : Data biasanya dibagi menjadi dua set, yaitu set pelatihan (training set) dan set pengujian (test set). Set pelatihan digunakan untuk membangun model, sementara set pengujian digunakan untuk mengevaluasi kinerja model.
4. Pemilihan Model : Algoritma machine learning yang sesuai dipilih berdasarkan masalah yang ingin diselesaikan. Beberapa jenis algoritma yang umum digunakan termasuk regresi, klasifikasi, clustering, dan lain-lain.
5. Pelatihan Model : Model dilatih menggunakan set pelatihan. Proses ini melibatkan penggunaan data untuk mengoptimalkan parameter model agar dapat membuat prediksi yang akurat.
6. Evaluasi Model : Model yang telah dilatih diuji menggunakan set pengujian untuk mengevaluasi kinerjanya. Metode evaluasi bisa berbeda-beda tergantung pada jenis masalah, misalnya akurasi, presisi, recall, F1-score, dan sebagainya.
7. Penyempurnaan Model : Berdasarkan hasil evaluasi, model dapat disesuaikan atau dioptimalkan lebih lanjut.
8. Prediksi : Setelah model berhasil dilatih dan dievaluasi, model dapat digunakan untuk membuat prediksi atau keputusan berdasarkan data baru.

Jenis-Jenis Machine Learning
1. Supervised Learning : Model dilatih menggunakan data yang berlabel, di mana input dan output yang benar diketahui. Contohnya termasuk regresi linear dan pohon keputusan.
2. Unsupervised Learning : Model dilatih menggunakan data yang tidak berlabel dan harus menemukan pola atau struktur dalam data. Contohnya termasuk clustering dan asosiasi.
3. Semi-supervised Learning : Kombinasi antara supervised dan unsupervised learning, di mana sebagian data diberi label dan sebagian lainnya tidak.
4. Reinforcement Learning: Model belajar melalui trial and error untuk mengambil tindakan yang memaksimalkan reward atau keuntungan dalam suatu lingkungan tertentu. Contohnya adalah algoritma yang digunakan dalam game AI dan robotika.

3) Machine learning memiliki beberapa macam taksonomi atau kategori yang mencakup berbagai pendekatan dan teknik yang digunakan untuk membangun model prediktif atau deskriptif dari data. Berikut adalah macam-macam takson machine learning secara detail:

1. Supervised Learning (Pembelajaran Terawasi)
Supervised learning melibatkan pelatihan model pada dataset berlabel, di mana input dipasangkan dengan output yang diharapkan. Ada dua kategori utama:

- Regression (Regresi) : Memodelkan hubungan antara input dan output numerik kontinu.
  - Linear Regression : Memodelkan hubungan linear antara input dan output.
  - Polynomial Regression : Memodelkan hubungan non-linear dengan menggunakan polinomial.
  - Ridge Regression : Menggunakan regularisasi untuk mencegah overfitting.
  - Lasso Regression : Menggunakan regularisasi L1 untuk mendorong sparseness dalam parameter.

- Classification (Klasifikasi): Memodelkan hubungan antara input dan output kategorikal.
  - Logistic Regression : Digunakan untuk masalah klasifikasi biner.
  - Support Vector Machines (SVM) : Mencari hyperplane optimal untuk memisahkan kelas-kelas.
  - Decision Trees : Menggunakan struktur pohon untuk memodelkan keputusan berdasarkan fitur.
  - Random Fore : Menggabungkan beberapa pohon keputusan untuk meningkatkan akurasi dan mengurangi overfitting.
  - k-Nearest Neighbors (k-NN) : Mengklasifikasikan berdasarkan mayoritas tetangga terdekat dalam ruang fitur.

2. Unsupervised Learning (Pembelajaran Tak Terawasi)
Unsupervised learning melibatkan pelatihan model pada dataset tanpa label, bertujuan menemukan pola atau struktur dalam data. Kategori utama meliputi:

- Clustering (Pengelompokan) : Mengelompokkan data ke dalam kelompok berdasarkan kemiripan.
  - k-Means Clustering : Mengelompokkan data menjadi k kelompok dengan meminimalkan jarak dalam kelompok.
  - Hierarchical Clustering : Membangun hierarki pengelompokan melalui penggabungan atau pemisahan bertahap.
  - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) : Mengelompokkan data berdasarkan kepadatan titik.

- Association (Asosiasi) : Menemukan aturan atau hubungan menarik antara fitur-fitur dalam data.
  - Apriori Algorithm : Mengidentifikasi itemset yang sering muncul dan aturan asosiasi di antara mereka.
  - Eclat Algorithm : Versi yang lebih efisien dari Apriori yang menggunakan pencarian kedalaman.

- Dimensionality Reduction (Pengurangan Dimensi): Mengurangi jumlah fitur dalam data sambil mempertahankan informasi penting.
  - Principal Component Analysis (PCA) : Mengubah data ke dalam komponen utama yang paling berkontribusi pada variasi.
  - t-Distributed Stochastic Neighbor Embedding (t-SNE) : Mengurangi dimensi untuk visualisasi dengan mempertahankan hubungan lokal.
  - Linear Discriminant Analysis (LDA) : Digunakan untuk pengurangan dimensi dan klasifikasi.

3. Semi-Supervised Learning (Pembelajaran Semi-Terawasi)
Semi-supervised learning menggunakan kombinasi data berlabel dan tidak berlabel untuk pelatihan. Pendekatan ini berguna ketika data berlabel sulit atau mahal didapat.

- Self-Training : Model dilatih secara bertahap dengan menambahkan prediksi berlabel dari data tidak berlabel ke set pelatihan.
- Co-Training : Dua model dilatih secara paralel dan saling memberikan prediksi dari data tidak berlabel.

4. Reinforcement Learning (Pembelajaran Penguatan)
Reinforcement learning melibatkan pelatihan agen yang belajar dengan berinteraksi dengan lingkungan dan mendapatkan reward atau penalti berdasarkan tindakan yang dilakukan. Tujuannya adalah untuk memaksimalkan reward total.

- Q-Learning: Algoritma yang mencari kebijakan optimal dengan menggunakan tabel nilai Q untuk menyimpan nilai dari setiap pasangan tindakan-keadaan.
- Deep Q-Networks (DQN) : Memperluas Q-Learning dengan menggunakan jaringan saraf dalam untuk mengaproksimasi nilai Q.
- Policy Gradient Methods : Langsung memodelkan dan mengoptimalkan kebijakan, bukan fungsi nilai.

5. Ensemble Learning
Ensemble learning menggabungkan prediksi dari beberapa model untuk meningkatkan kinerja dan keakuratan.

- Bagging (Bootstrap Aggregating): Meningkatkan stabilitas dan akurasi model dengan menggabungkan beberapa model yang dilatih pada subset data yang berbeda. Contohnya adalah Random Forest.
- Boosting : Menggabungkan model secara berurutan, di mana setiap model baru berusaha untuk memperbaiki kesalahan dari model sebelumnya. Contohnya adalah AdaBoost dan Gradient Boosting.
- Stacking : Menggabungkan prediksi dari beberapa model dasar dan menggunakan model meta untuk membuat prediksi akhir.

6. Neural Networks dan Deep Learning
Deep learning adalah sub-bidang machine learning yang menggunakan jaringan saraf tiruan (artificial neural networks) dengan banyak lapisan (deep) untuk memodelkan data yang sangat kompleks.

- Convolutional Neural Networks (CNNs)**: Jaringan saraf yang sangat efektif untuk tugas pengenalan gambar dan analisis data grid-like.
- Recurrent Neural Networks (RNNs)**: Jaringan saraf yang cocok untuk data urutan, seperti teks atau data waktu-nyata.
- Generative Adversarial Networks (GANs)**: Model yang terdiri dari dua jaringan (generator dan discriminator) yang berkompetisi untuk menghasilkan data yang mirip dengan data asli.
- Autoencoders**: Jaringan saraf yang digunakan untuk pengurangan dimensi dan deteksi anomali.

    a) Berapa rata-rata mahasiswa datang pada minggu ini?  
    b) Kapan biaya tertinggi terjadi?  
    c) Hari apa biaya lebih dari 110000?  
    d) Siapa yang paling banyak datang ke kampus?  
    e) Siapa yang datang pada hari minggu?  
    f) Berapa biaya tertinggi dan terendah?  
    g) Berapa frekuensi datang tertinggi dan terendah?  

````python
# Data set
data_mahasiswa = {
    "Ani": {"Senin": 2, "Biaya": 30000},
    "Budi": {"Selasa": 3, "Biaya": 35000},
    "Joni": {"Jumat": 2, "Biaya": 20000},
    "Jono": {"Rabu": 4, "Biaya": 25000},
    "Lono": {"Kamis": 1, "Biaya": 15000},
    "Minggu": {"Sabtu": 5}
}

# Menghitung total biaya per mahasiswa
# Menempatkan perulangan setelah komentar
for mahasiswa in data_mahasiswa:
    if mahasiswa != "Minggu":
        data_mahasiswa[mahasiswa]["Total_Biaya"] = data_mahasiswa[mahasiswa]["Biaya"] * data_mahasiswa[mahasiswa][list(data_mahasiswa[mahasiswa].keys())[0]]

# a) Rata-rata mahasiswa datang pada minggu ini
total_datang = sum([sum(data_mahasiswa[mahasiswa].values()) for mahasiswa in data_mahasiswa if mahasiswa != "Minggu"])
rata_rata_datang = total_datang / len(data_mahasiswa)

# b) Kapan biaya tertinggi terjadi
hari_biaya_tertinggi = max(data_mahasiswa, key=lambda x: data_mahasiswa[x].get("Total_Biaya", 0))

# c) Hari apa biaya lebih dari 110000
hari_lebih_dari_110000 = [hari for hari in data_mahasiswa if data_mahasiswa.get(hari, {}).get("Total_Biaya", 0) > 110000]

# d) Siapa yang paling banyak datang ke kampus
mahasiswa_terbanyak = max(data_mahasiswa, key=lambda x: sum(data_mahasiswa[x].values()) if x != "Minggu" else 0)

# e) Siapa yang datang pada hari Minggu
mahasiswa_hari_minggu = [mahasiswa for mahasiswa in data_mahasiswa if "Minggu" in data_mahasiswa[mahasiswa]]

# f) Biaya tertinggi dan terendah
biaya_tertinggi = max([data_mahasiswa[mahasiswa].get("Total_Biaya", 0) for mahasiswa in data_mahasiswa if mahasiswa != "Minggu"])
biaya_terendah = min([data_mahasiswa[mahasiswa].get("Total_Biaya", float('inf')) for mahasiswa in data_mahasiswa if mahasiswa != "Minggu"])

# g) Frekuensi datang tertinggi dan terendah
frekuensi_tertinggi = max([sum(data_mahasiswa[mahasiswa].values()) for mahasiswa in data_mahasiswa if mahasiswa != "Minggu"])
frekuensi_terendah = min([sum(data_mahasiswa[mahasiswa].values()) for mahasiswa in data_mahasiswa if mahasiswa != "Minggu"])

# Menampilkan hasil
print("a) Rata-rata mahasiswa datang pada minggu ini:", rata_rata_datang)
print("b) Kapan biaya tertinggi terjadi:", hari_biaya_tertinggi)
print("c) Hari apa biaya lebih dari 110000:", hari_lebih_dari_110000)
print("d) Siapa yang paling banyak datang ke kampus:", mahasiswa_terbanyak)
print("e) Siapa yang datang pada hari Minggu:", mahasiswa_hari_minggu)
print("f) Biaya tertinggi:", biaya_tertinggi, "Biaya terendah:", biaya_terendah)
print("g) Frekuensi datang tertinggi:", frekuensi_tertinggi, "Frekuensi datang terendah:", frekuensi_terendah)
````
